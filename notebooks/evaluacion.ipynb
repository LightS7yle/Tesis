```python
df_bert = pd.read_csv("../resultados/evaluacion_con_bert.csv")
df_franq = pd.read_csv("../resultados/evaluacion_rag_con_franq.csv")
df_rouge = pd.read_csv("../resultados/evaluacion_con_rouge.csv")
```

```python
df_bert.head()
```

```python
df_rouge.head()
```

```python
df_franq.head()
```

```python
df_comb = pd.concat([
    df_bert[["question"]],
    df_bert[["PRECISION"]],
    df_bert[["RECALL"]],
    df_bert[["F1"]],
    df_franq[["0"]],
    df_rouge[["ROUGE-1"]],
    df_rouge[["ROUGE-2"]],
    df_rouge[["ROUGE-L"]]
], axis=1)
```

```python
df_comb.rename(columns={"0": "FRANQ"}, inplace=True)
```

```python
df_comb.head()
```

```python
mean_P=[]
mean_R=[]
mean_F1=[]
mean_FRANQ=[]
mean_ROUGE_1=[]
mean_ROUGE_2=[]
mean_ROUGE_L=[]
mean_P.append(df_comb["PRECISION"].mean().round(4))
mean_R.append(df_comb["RECALL"].mean().round(3))
mean_F1.append(df_comb["F1"].mean().round(3))
mean_FRANQ.append(df_comb["FRANQ"].mean().round(3))
mean_ROUGE_1.append(df_comb["ROUGE-1"].mean().round(3))
mean_ROUGE_2.append(df_comb["ROUGE-2"].mean().round(3))
mean_ROUGE_L.append(df_comb["ROUGE-L"].mean().round(3))

df_mean = pd.DataFrame({
    "PRECISION": mean_P,
    "RECALL": mean_R,
    "F1": mean_F1,
    "FRANQ": mean_FRANQ,
    "ROUGE-1": mean_ROUGE_1,
    "ROUGE-2": mean_ROUGE_2,
    "ROUGE-L": mean_ROUGE_L,
  
})
```
